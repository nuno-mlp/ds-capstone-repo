We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.


```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=10) # Create a Map with Folium and Leaflet.js
```

We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example,


```python
# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center')) # radius (float) – Radius of the circle, in meters.
# Create a marker at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker( # folium.Marker() é a forma padrão e mais usada. Marker é importado diretamente do módulo principal folium. Marker também existe dentro do submódulo folium.map então é tecnicamente equivalente.
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon( # Substitui o ícone padrão (pin vermelho) por um ícone HTML customizado.
        icon_size=(20,20), # Tamanho do ícone (largura, altura) em pixels.
        icon_anchor=(0,0), # Define o ponto do ícone que corresponde à coordenada no mapa. (0,0) significa canto superior esquerdo do div.      
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC', # Aqui definimos o conteúdo HTML do ícone.
        # <div> = contêiner HTML do ícone.
        # style="font-size: 12; color:#d35400;" = cor laranja e tamanho de fonte 12px.
        # <b>%s</b> = texto em negrito, que neste caso é "NASA JSC" (substituído pelo % 'NASA JSC').
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```


| Parâmetro       | Tipo   | Valor no teu código | Função                                                                                                           |
| --------------- | ------ | ------------------- | ---------------------------------------------------------------------------------------------------------------- |
| `position`      | `str`  | `'topright'`        | Onde o widget aparece no mapa. Opções: `'topleft'`, `'topright'`, `'bottomleft'`, `'bottomright'`.               |
| `separator`     | `str`  | `' Long: '`         | Texto que separa latitude de longitude no display. Ex.: `Lat: 41.7 Long: -8.83`.                                 |
| `empty_string`  | `str`  | `'NaN'`             | O que mostrar quando a posição ainda não está disponível.                                                        |
| `lng_first`     | `bool` | `False`             | Se `True`, mostra **longitude primeiro** (`Lng: ... Lat: ...`). Se `False`, mostra latitude primeiro.            |
| `num_digits`    | `int`  | `5`                 | Número de casas decimais para exibir na latitude e longitude.                                                    |
| `prefix`        | `str`  | `'Lat:'`            | Texto que aparece antes do valor da **latitude**.                                                                |
| `lat_formatter` | função | `None` (comentado)  | Permite passar uma função customizada para formatar o valor da latitude. Ex.: arredondar, adicionar símbolo `°`. |
| `lng_formatter` | função | `None` (comentado)  | Igual ao anterior, mas para longitude.                                                                           |


Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to calculate the distance to the launch site.


```python
from math import sin, cos, sqrt, atan2, radians
```

| Função        | O que faz                                                              | Input                         | Unidade/Retorno         | Exemplo                       |
| ------------- | ---------------------------------------------------------------------- | ----------------------------- | ----------------------- | ----------------------------- |
| `sin(x)`      | Calcula o seno do ângulo                                               | Radianos                      | Retorna entre -1 e 1    | `sin(radians(90)) → 1.0`      |
| `cos(x)`      | Calcula o cosseno do ângulo                                            | Radianos                      | Retorna entre -1 e 1    | `cos(radians(0)) → 1.0`       |
| `sqrt(x)`     | Calcula a raiz quadrada                                                | Número real ≥ 0               | Retorna número real ≥ 0 | `sqrt(16) → 4.0`              |
| `atan2(y, x)` | Calcula o ângulo cuja tangente é y/x, considerando o quadrante correto | Dois números reais (`y`, `x`) | Radianos, entre -π e π  | `degrees(atan2(1,1)) → 45.0°` |
| `radians(x)`  | Converte graus em radianos                                             | Graus                         | Radianos                | `radians(180) → 3.14159`      |


```python
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
